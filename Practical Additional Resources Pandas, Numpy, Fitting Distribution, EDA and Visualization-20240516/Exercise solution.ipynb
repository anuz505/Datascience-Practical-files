```python
new_df = df.fillna({
        
        'Annual Income (k$)': 0,
        'Gender': 'No Gender'
    })
new_df.isnull().sum()
new_df.head()
```

```python
num = [col for col in new_df.columns if new_df[col].dtypes != 'O']
new_df[num].head()
```
