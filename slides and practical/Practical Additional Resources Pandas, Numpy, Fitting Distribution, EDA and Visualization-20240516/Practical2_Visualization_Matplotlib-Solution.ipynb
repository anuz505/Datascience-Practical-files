### Example

A simple figure with MATLAB-like plotting API:

```python
#Create a dataset
x = np.linspace(0, 20, 10)
y = x ** 2
print((x,y))
```

```python
plt.figure()
plt.plot(x, y, 'r*--')
plt.xlabel('x')
plt.ylabel('square of x')
plt.title('This is a demo plot')
plt.show()

```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'y^:')
plt.subplot(1,2,2)
plt.plot(y, x, 'b*--');
```

```python
## change the first plot to marker green diamond and second plot to marker yellow circle
plt.subplot(1,2,1)
plt.plot(x, y, 'g^-')
plt.subplot(1,2,2)
plt.plot(y, x, 'yo--');
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2)
axes[0].plot( x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

## Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  

```

## Pie  chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%')

plt.axis('equal')
plt.show()
```
